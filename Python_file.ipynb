```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == '+':
    print("Result:", a + b)
elif op == '-':
    print("Result:", a - b)
elif op == '*':
    print("Result:", a * b)
elif op == '/':
    if b != 0:
        print("Result:", a / b)
    else:
        print("Cannot divide by zero!")
else:
    print("Invalid operator")
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
print("Is prime?", is_prime(17))
```

```python

def get_element(lst, index):
    try:
        return lst[index]
    except IndexError:
        return "Error: Index out of range."

my_list = [10, 20, 30]
print(get_element(my_list, 2))  # valid
print(get_element(my_list, 5))  # invalid
```

```python

try:
    with open("Hello.txt", "r") as f:
        lines = f.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    print("Lines:", line_count)
    print("Words:", word_count)
    print("Characters:", char_count)
except FileNotFoundError:
    print("File not found.")
```

```python

try:
    with open("Hello.txt", "r") as f:
        lines = f.readlines()
    with open("Hello_no_blanks.txt", "w") as f:
        for line in lines:
            if line.strip():  # non-blank line
                f.write(line)
    print("Blank lines removed and saved to Hello_no_blanks.txt")
except FileNotFoundError:
    print("File not found.")
```
